```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({'A': [1,2,np.nan, 4], 
                   'B': [5, np.nan, np.nan, 8], 
                   'C': [10, 20, 30, 40]})
df
```

```python
df.dropna()
```

```python
df.dropna(axis=1)
```

```python
df.dropna(axis=1, thresh=2) # thresh requires the number of non-null values; can base this off some percentage
```

```python
df.dropna(axis=1, thresh=3)
```

```python
df.dropna(axis=0, thresh=3)
```

```python
df.fillna(value='FILL_VALUE') # NOTE: values are converted to ints
```

```python
df.fillna(value=0)
```

```python
df['A'] = df['A'].fillna(value=0)
df
```

```python
df = pd.DataFrame({'A': [1,2,np.nan, 4], 
                   'B': [5, np.nan, np.nan, 8], 
                   'C': [10, 20, 30, 40]})
df
```

```python
# Or fill in missing values with the mean for the entire data frame:
df = df.fillna(df.mean())
df
```
